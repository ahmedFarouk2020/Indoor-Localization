## 1. Delete unrelated data

```python
import pandas as pd

filename = '5_4'

stairs_left = pd.read_csv(f'./dataset-separated/5/{filename}.csv')
stairs_left.head()
```

```python
stairs_left = stairs_left.drop(0)
stairs_left = stairs_left.drop(1)
stairs_left.head()
```

```python
# change columns labels
stairs_left.rename ( {"#!": "APs", "Workbook:": "Strength"}, axis=1, inplace =True)
stairs_left.head()
```

```python
# delete the unwanted columns [custom, custom.1]
stairs_left.drop(["Custom", "Custom.1"], axis=1, inplace=True)
stairs_left.info()
```

```python
# delete undesired APs

for i in range(2,207):
    if stairs_left.loc[i][0] in ['STUDBME1', 'STUDBME2', 'CMP_LAB1','CMP_LAB2','CMP_LAB3','CMP_LAB4','s']:
        continue
    else:
        stairs_left = stairs_left.drop(i)
```

```python
stairs_left
```

```python
stairs_left.info()
```

```python
# create a new df
new_df = pd.DataFrame(columns =['STUDBME1', 'STUDBME2', "CMP_LAB1", "CMP_LAB2", "CMP_LAB3", "CMP_LAB4"], dtype=int)


new_df
```

```python
STUDBME1 = 0
STUDBME2 = 0
CMP_LAB1 = 0
CMP_LAB2 = 0
CMP_LAB3 = 0
CMP_LAB4 = 0

record_id = 0
for i in range(74):
    if stairs_left.iloc[i][0] == 'STUDBME1':
        STUDBME1 = stairs_left.iloc[i][1]
    elif stairs_left.iloc[i][0] == 'STUDBME2':
        STUDBME2 = stairs_left.iloc[i][1]
    elif stairs_left.iloc[i][0] == 'CMP_LAB1':
        CMP_LAB1 = stairs_left.iloc[i][1]
    elif stairs_left.iloc[i][0] == 'CMP_LAB2':
        CMP_LAB2 = stairs_left.iloc[i][1]
    elif stairs_left.iloc[i][0] == 'CMP_LAB3':
        CMP_LAB3 = stairs_left.iloc[i][1]
    elif stairs_left.iloc[i][0] == 'CMP_LAB4':
        CMP_LAB4 = stairs_left.iloc[i][1]
    elif stairs_left.iloc[i][0] == 's':
        # insert the record in df
        new_df.loc[record_id] = [STUDBME1,STUDBME2,CMP_LAB1,CMP_LAB2,CMP_LAB3,CMP_LAB4]
        record_id +=1
    else:
        pass
    
    
    
new_df.head()
```

```python
new_df["target"] = 5
new_df.head()
```

```python
# save the data to a file
filename = "5_4"
new_df.to_csv(f"./ready-to-model/{filename}.csv", index=False)
print(f"./ready-to-model/{filename}.csv")
```
