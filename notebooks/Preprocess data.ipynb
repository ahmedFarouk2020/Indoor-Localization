## Configure prerequisite

```python
import pandas as pd
import numpy as np
import os

directory = "data-set/"
files_list = os.listdir(directory)

#print(files_list)

all_data = pd.DataFrame(columns=['STUDBME2', 'CMP_LAB3','Sbme-Staff','Farouk', 'youssef', 'target'], dtype=np.int8)
files_list = ['1.csv', '10.csv', '11.csv', '12.csv', '2.csv', '3.csv', '8.csv', '9.csv']
files_list
```

## Read & Collect all data files into one file

```python
for csv in files_list:
    # load a csv file
    file_csv = pd.read_csv(f'{directory}{csv}')
    
    file_csv.astype(np.int8)
    
    # add target column to file
    file_csv['target'] = int(csv.split('.')[0])
    print(file_csv)
    # append formatted data
    all_data = all_data.append(file_csv)
    
all_data
```

## Save the data file

```python
all_data.to_csv(f'{directory}all-data.csv')
```
